Gradient descent- iterating algorithm used to minimise a function by adjusting the parameters in the steepest distance distance(like a negative gradient).

```python
import numpy as np
import matplotlib.pyplot as plt

def fun(x):
    return np.cos(2 * np.pi * x) + x**2

def gra(x):
    return -2 * np.pi * np.sin(2 * np.pi * x) + 2 * x

def f_min(start_x, step=0.05, max_iter=100, tol=1e-6):
    x = start_x
    path = [x]

    for _ in range(max_iter):
        update = x - step * gra(x)
        path.append(update)
        if abs(update - x) < tol:
            break
        x = update

    return x, path

best_x, journey = f_min(1.0)

x_vals = np.linspace(-2, 2, 400)
plt.figure(figsize=(10, 6))
plt.plot(x_vals, fun(x_vals), 'b-', label="f(x) = cos(2πx) + x²")
plt.scatter(journey, [fun(x) for x in journey], c='r', s=50, label="Gradient Descent Steps")
plt.plot(journey, [fun(x) for x in journey], 'r--', alpha=0.3, label="Optimization Path")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Gradient Descent Journey")
plt.legend()
plt.grid(True)
plt.show()

print(f"Found minimum at x = {best_x:.4f} (f(x) = {fun(best_x):.4f})")
```
